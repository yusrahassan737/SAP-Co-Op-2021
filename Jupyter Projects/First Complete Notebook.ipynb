### Cleaning and Display
From the last step, it looks like there are no duplicate values and the head shows that the df.dtypes are in fact the correct data types -both good and clean. On the other hand, we found that there were only 164 entries in the Calories column although there are 169 lines of entries. This means that the Calories column has 5 null values.

The method I will use to clean up the null values is to simply drop those lines. It is quick, simple and authentic as values do not have to be recreated based on an average. The dropna method usually returns a new data frame without the incomplete lines, but in this case, it doesn't make a distinguishable difference in the visualisations, so I will have it replace the old one.

```python
df.dropna(inplace = True)
```
